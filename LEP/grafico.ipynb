```python
import ROOT
```

```python
entrada = ROOT.TFile("data/root/data.root",'read')
tree = entrada.Get("tree")
```

```python
print(tree.GetNbranches())
```

```python
for nome in tree.GetListOfBranches():
    print(nome)
```

```python
c= ROOT.TCanvas()
c.SetTickx() 
c.SetTicky()
c.SetGridx()
c.SetGridy()
histogram = ROOT.TH1F("histograma_phi", "Distribuicao de phi para abs(id)==321 e abs(id)==211;\pi ;Contagem", 100, -ROOT.Math.Pi(),ROOT.Math.Pi())
tree.Draw("phi >> histograma_phi", "abs(id) == 321 || abs(id) == 211")


histogram.GetXaxis().SetTitleSize(0.05) # Tamanho legenda X
histogram.GetYaxis().SetTitleSize(0.05)
histogram.GetYaxis().SetRangeUser(0, 200)
histogram.SetFillColor(ROOT.kBlue-9)
# Crie uma janela de visualização para o histograma
canvas = ROOT.TCanvas("canvas", "Canvas", 800, 600)
histogram.Draw()
canvas.Update()
```

```python
c.Draw()
```
